```python
import matplotlib.pyplot as plt
import numpy as np

base = np.load('baseLoss.npy')
L1 = np.load('L1Loss.npy')
EI = np.load('EILoss.npy')

steps = np.arange(0, 2000, 100)
plt.figure(figsize=(10,5))
plt.title("Learning Curve")
plt.plot(steps,base, label='Standard Leaky RNN',color='green')
plt.plot(steps,EI, label='EI Leaky RNN',color='navy')
plt.plot(steps,L1, label='L1 and L2 Regularised Leaky RNN',color='brown')

plt.xlabel("Steps")
plt.xlim(0,2000)
plt.ylim(0,0.8)
plt.ylabel("Loss")
plt.legend()
plt.show()
```
